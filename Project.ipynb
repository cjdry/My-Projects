```python
plt.figure(figsize=(12,8)) 
sns.heatmap(d1.corr(), cmap='viridis') 
```

```python
#saleprice correlation matrix
k=10 #number of variables for heatmap
cols=d1.corr().nlargest(k,'SalePrice')['SalePrice'].index
cm=d1[cols].corr()
plt.figure(figsize=(10,6))
sns.heatmap(cm,annot=True,cmap='viridis') 
```

```python
cols=['SalePrice','OverallQual','GrLivArea','GarageCars','TotalBsmtSF','FullBath','YearBuilt']
sns.pairplot(d1[cols]) 
```

```python
c_f_c=list(set(d1.columns)-set(d1._get_numeric_data().columns))
c_f_c
```

```python
print(d1.shape)
d1.dropna(inplace=False)
print(d1.shape) 
```

```python
sns.countplot(x='SaleCondition', hue='Neighborhood',data=d1)
plt.show() 
```

```python
Combined_Variable= n_f_c + c_f_c
print(Combined_Variable) 
```

```python

facet=None
for i in range(0,len(Combined_Variable),2):
    plt.figure(figsize=(10,4))
    plt.subplot(121)
    sns.boxplot(facet,Combined_Variable[i],data=d1)
    plt.subplot(122)
    sns.boxplot(facet,Combined_Variable[i+1],data=d1)
    plt.tight_layout()
    plt.show() 
```
