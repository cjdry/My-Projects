```python
le_x2=LabelEncoder()
le_x1=LabelEncoder()
le_x3=LabelEncoder()
le_x0=LabelEncoder()
le_x8=LabelEncoder()
le_x4=LabelEncoder()
le_x5=LabelEncoder()
le_x6=LabelEncoder()
train_data['X0']=le_x0.fit_transform(train_data['X0'])
train_data['X2']=le_x2.fit_transform(train_data['X2'])
train_data['X1']=le_x1.fit_transform(train_data['X1'])
train_data['X3']=le_x3.fit_transform(train_data['X3'])
train_data['X4']=le_x4.fit_transform(train_data['X4'])
train_data['X5']=le_x5.fit_transform(train_data['X5'])
train_data['X6']=le_x6.fit_transform(train_data['X6'])
train_data['X8']=le_x8.fit_transform(train_data['X8'])

test_data['X0']=le_x0.fit_transform(test_data['X0'])
test_data['X2']=le_x2.fit_transform(test_data['X2'])
test_data['X1']=le_x1.fit_transform(test_data['X1'])
test_data['X3']=le_x3.fit_transform(test_data['X3'])
test_data['X4']=le_x4.fit_transform(test_data['X4'])
test_data['X5']=le_x5.fit_transform(test_data['X5'])
test_data['X6']=le_x6.fit_transform(test_data['X6'])
test_data['X8']=le_x8.fit_transform(test_data['X8'])
train_data.head()
```

```python
test_data.head()
```

```python
new_train_data=pd.concat([train_data],axis=1)
new_train_data.head()
```

```python
new_test_data=pd.concat([test_data],axis=1)
new_test_data.head()
```

```python
features=new_train_data.drop(['y'],axis=1)
target=new_train_data[['y']]
X_train,X_test,y_train,y_test=train_test_split(features,target,train_size=0.85,random_state=5)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
my_model=LinearRegression()
my_model.fit(X_train,y_train)
predictions=my_model.predict(X_test)
predictions_train=my_model.predict(X_train)
print('MSE:',mean_squared_error(predictions,y_test))
print('RMSE:',np.sqrt(mean_squared_error(predictions,y_test)))
print('r2_score on test data:',r2_score(predictions,y_test))
print('r2_score on train data:',r2_score(predictions_train,y_train))
```

```python
my_model=Ridge(alpha=5)
my_model.fit(X_train,y_train)
predictions=my_model.predict(X_test)
print('MSE:',mean_squared_error(predictions,y_test))
print('RMSE',np.sqrt(mean_squared_error(predictions,y_test)))
print('r2_score:',r2_score(predictions,y_test))
```

```python
my_model=Lasso()
my_model.fit(X_train,y_train)
predcitions=my_model.predict(X_test)
print('MSE:',mean_squared_error(predictions,y_test))
print('RMSE',np.sqrt(mean_squared_error(predictions,y_test)))
print('r2_score:',r2_score(predictions,y_test))
```

```python
my_model=ElasticNet()
my_model.fit(X_train,y_train)
predictions=my_model.predict(X_test)
print('MSE:',mean_squared_error(predictions,y_test))
print('RMSE',np.sqrt(mean_squared_error(predictions,y_test)))
```
