```python
# Importing the required Library

import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, MaxPooling2D, Flatten, Conv2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
```

```python
train_image_dir = 'C:/Users/project_3/data/train/'
test_image_dir = 'C:/Users/project_3/data/test/'
```

```python


train_generator = ImageDataGenerator(rescale=1.0/255.,
                                    rotation_range=30,
                                    zoom_range=0.2,
                                    horizontal_flip=True)

# For Test Data
test_genrator = ImageDataGenerator(rescale=1.0/255.)
```

```python


batch_size = 32

train_datasets = train_generator.flow_from_directory(train_image_dir, 
                                                     target_size=(64, 64),
                                                     color_mode="rgb", 
                                                     batch_size=batch_size,
                                                     class_mode="categorical",
                                                     subset='training',
                                                     shuffle=True)


test_datasets = test_genrator.flow_from_directory(test_image_dir,
                                                target_size=(64, 64),
                                                color_mode="rgb",
                                                batch_size=1,
                                                class_mode='categorical',
                                                shuffle=True)
```

```python
# Defining the Architecture of CNN Model

fc_size = 32
model = Sequential()

# 1st Convolution Layer
model.add(Conv2D(32, (5,5), input_shape=(64,64,3), activation='relu', padding='same', strides=(1,1)))
model.add(MaxPooling2D(pool_size=(2,2), strides=(2,2)))

# 2nd Convolution Layer
model.add(Conv2D(64,(5,5), input_shape=(64,64,3), activation='relu', padding='same', strides=(1,1)))
model.add(MaxPooling2D(pool_size=(2,2), strides=(2,2)))

model.add(Flatten())

model.add(Dense(units=fc_size, activation='relu'))

model.add(Dropout(0.4))

#
model.add(Dense(2, activation='softmax'))
```

```python
model.summary()
```

```python
# Compile the model

model.compile(loss='categorical_crossentropy', optimizer=Adam(learning_rate=0.01), metrics=['accuracy'])
```

```python

history_100_epochs = model.fit(train_datasets,
                   validation_data = test_datasets,
                   epochs=100,
                   verbose=1)
```

```python
score = model.evaluate(test_datasets, verbose=1)
```

```python
print("Test Score: ", score[1]*100)
print("Test Loss: ", score[0]*100)
```

```python
# Prediction

pred = model.predict(test_datasets)

y_class = pred.argmax(axis=-1)
y_class
```
